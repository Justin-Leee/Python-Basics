- 작성한 코드의 정답 여부는, input으로 넣은 문장을 제대로 분류해서 실행하는지 확인함으로써 validate될 것이다.

````python
problem_4 = input('hey wats up')
ㅎ = []

def autoMmmmultiply(data, kimbab):

  # Write further code
  if kimbab[kimbab.index('/')+1:] == 'apnd':
    tmp = (kimbab[:kimbab.index('/')]).split(',')

    data.append([int(bunny) for bunny in tmp])

  elif kimbab[kimbab.index('/')+1:] == 'ext':
    tmp = (kimbab[:kimbab.index('/')]).split(',')
    data.extend([int(bunny) for bunny in tmp])
  
  
  return data # Do not change this line 
autoMmmmultiply(ㅎ,problem_4)
````
