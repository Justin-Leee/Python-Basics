## Problem 1. Create a function named ```multiplicationTable``` which takes one parameter and prints out all the possible multiplications with it and 1 to 9. The number will be an user input.

For example, if you input ```6```, the function will print out the following lines;

```python
6 x 1 = 6
6 x 2 = 12
6 x 3 = 18
...

6 x 8 = 48
6 x 9 = 54
```

````python
def multiplicationTable(mango):
  zero = 0
  while zero < 10:
    zero += 1
    print(f'{mango} x {zero} = {int(mango)*zero}')
multiplicationTable(input())
````

## HW 1. Create a function named ```findMax``` which takes unlimited number of integers and returns the maximum among them. 

Note : Using a max() function or any other functions which help you automatically find the maximum is not allowed.

Hint : Do you remember what **kwargs in function is?

````python
def findMax(*args):

  a = args[0]
  for i in args:
    if i > a:
      a = i
  return a





# All the code lines blow this line must be runnable.

print(findMax(1,3,2,5,634,7,4,3235,2,34,67,7))
````
